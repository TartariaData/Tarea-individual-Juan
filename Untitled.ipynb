```python
!pip install geopandas
import os
import geopandas as gpd

# Cargar archivos shapefile
countries = gpd.read_file(os.path.join("maps", "World_Countries", "World_Countries.shp"))
rivers = gpd.read_file(os.path.join("maps", "World_Hydrography", "World_Hydrography.shp"))
cities = gpd.read_file(os.path.join("maps", "World_Cities", "World_Cities.shp"))

# Verificar la carga de datos
print(countries.head())
print(rivers.head())
print(cities.head())
```

```python
import pandas as pd

# Cargar datos de fragilidad
fragilityCiaLink = "https://github.com/CienciaDeDatosEspacial/merging/raw/main/FragilityCia_isos.csv"
fragilityCia = pd.read_csv(fragilityCiaLink)

# Ajustar nombres de columnas para la fusión
countries['COUNTRY'] = countries.COUNTRY.str.upper()
onlyFragilCia = set(fragilityCia.Country) - set(countries.COUNTRY)
onlyMap = set(countries.COUNTRY) - set(fragilityCia.Country)

from thefuzz import process
try1 = {country: process.extractOne(country, onlyMap)[0] for country in sorted(onlyFragilCia) if process.extractOne(country, onlyMap)[1] >= 90}
fragilityCia.replace(to_replace={'Country': try1}, inplace=True)
manualChanges = {'SWAZILAND': 'ESWATINI', 'LAOS': "LAO PEOPLE'S DEMOCRATIC REPUBLIC (THE)"}
countries.replace(to_replace={'COUNTRY': manualChanges}, inplace=True)

# Verificar los cambios manuales
print(manualChanges)
```

```python
# Fusionar dataframes
theMapAndData = countries.merge(fragilityCia, left_on='COUNTRY', right_on='Country')
theMapAndData.drop(columns=['Country'], inplace=True)

# Verificar la fusión
print(theMapAndData.head())
print(theMapAndData.columns)
```

```python
from sklearn.preprocessing import QuantileTransformer

# Transformar y discretizar variables `co2` y `ForestRev_gdp`
DataNames = ['co2', 'ForestRev_gdp']
scaler = QuantileTransformer(n_quantiles=99, random_state=0, output_distribution='normal')
QtScaled_data = scaler.fit_transform(theMapAndData[DataNames])

theMapAndData['co2_Qt'] = QtScaled_data[:, 0]
theMapAndData['ForestRev_gdp_Qt'] = QtScaled_data[:, 1]

# Verificar la transformación y discretización
print(theMapAndData[['co2', 'co2_Qt', 'ForestRev_gdp', 'ForestRev_gdp_Qt']].head())
```

```python
!pip install mapclassify
```

```python
import mapclassify
import numpy as np

# Discretizar usando Jenks-Caspall
K = 5
theVar_co2 = theMapAndData.co2_Qt
jc5_co2 = mapclassify.JenksCaspall(theVar_co2, k=K)

theVar_forest = theMapAndData.ForestRev_gdp_Qt
jc5_forest = mapclassify.JenksCaspall(theVar_forest, k=K)

theMapAndData['co2_Qt_jc5'] = jc5_co2.yb
theMapAndData['forest_Qt_jc5'] = jc5_forest.yb

# Verificar la discretización
print(theMapAndData[['co2_Qt', 'co2_Qt_jc5', 'ForestRev_gdp_Qt', 'forest_Qt_jc5']].head())
```

```python
!pip install matplotlib
import matplotlib.pyplot as plt

def plot_map(data, column, title, cmap='viridis'):
    f, ax = plt.subplots(1, figsize=(10, 10))
    data.plot(column=column, cmap=cmap, categorical=True, edgecolor='white', linewidth=0.5, alpha=1, legend=True, legend_kwds={'loc': "lower left"}, ax=ax)
    ax.set_axis_off()
    ax.set_title(title)
    plt.show()

# Crear mapas para `co2` y `ForestRev_gdp`
plot_map(theMapAndData, 'co2_Qt_jc5', 'CO2 Emissions (Discretized)')
plot_map(theMapAndData, 'forest_Qt_jc5', 'Forest Revenue (Discretized)')
```

```python
# Crear datos adicionales de ejemplo
additional_data = pd.DataFrame({
    'Country': ['UNITED STATES', 'CANADA', 'MEXICO', 'BRAZIL', 'ARGENTINA'],
    'new_variable': [100, 200, 300, 400, 500]
})

# Suponiendo que la nueva variable es `new_variable` y tiene una columna `Country` para fusionar
additional_data['Country'] = additional_data['Country'].str.upper()
theMapAndData = theMapAndData.merge(additional_data, left_on='COUNTRY', right_on='Country', how='left')

# Verificar la fusión
print(theMapAndData.head())
print(theMapAndData.columns)
```

```python
!pip install scikit-learn
from sklearn.preprocessing import QuantileTransformer

# Transformar y discretizar la nueva variable
new_variable_name = 'new_variable'
scaler = QuantileTransformer(n_quantiles=99, random_state=0, output_distribution='normal')
QtScaled_new_variable = scaler.fit_transform(theMapAndData[[new_variable_name]].fillna(0))
theMapAndData[new_variable_name + '_Qt'] = QtScaled_new_variable

# Discretizar la nueva variable usando Jenks-Caspall
import mapclassify

K = 5
theVar_new = theMapAndData[new_variable_name + '_Qt']
jc5_new = mapclassify.JenksCaspall(theVar_new, k=K)
theMapAndData[new_variable_name + '_Qt_jc5'] = jc5_new.yb

# Verificar la discretización
print(theMapAndData[[new_variable_name, new_variable_name + '_Qt', new_variable_name + '_Qt_jc5']].head())
```

```python
import matplotlib.pyplot as plt

def plot_map(data, column, title, cmap='viridis'):
    f, ax = plt.subplots(1, figsize=(10, 10))
    data.plot(column=column, cmap=cmap, categorical=True, edgecolor='white', linewidth=0.5, alpha=1, legend=True, legend_kwds={'loc': "lower left"}, ax=ax)
    ax.set_axis_off()
    ax.set_title(title)
    plt.show()

# Crear mapa para la nueva variable
plot_map(theMapAndData, new_variable_name + '_Qt_jc5', 'New Variable (Discretized)', cmap='OrRd')
```

```python
# Eliminar la columna duplicada 'Country'
theMapAndData.drop(columns=['Country'], inplace=True)

# Intentar guardar nuevamente en formato GPKG
theMapAndData.to_file(os.path.join("maps", "worldMaps.gpkg"), layer='indicators_with_new_variable', driver="GPKG")
```
